```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
```

```python
#importing data
data=pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
```

```python
data.info()
```

```python
data.Date = pd.to_datetime(data.Date)
```

```python
data.dtypes
```

```python
data['Location'].fillna(data['Location'].mode()[0], inplace=True)
data['Operator'].fillna(data['Operator'].mode()[0], inplace=True)
data['Route'].fillna('Unknown', inplace=True)
data['Type'].fillna(data['Type'].mode()[0], inplace=True)
data['Registration'].fillna(data['Registration'].mode()[0], inplace=True)
data['cn/In'].fillna(data['cn/In'].mode()[0], inplace=True)
data['Aboard'].fillna(data['Aboard'].mean(), inplace=True)
data['Fatalities'].fillna(data['Fatalities'][0], inplace=True)
data['Ground'].fillna(data['Ground'].mean(), inplace=True)
data['Summary'].fillna(data['Summary'].mode()[0], inplace=True)
# Replace missing time values with midnight (00:00:00)
default_time = dt.time(0, 0)
data['Time'].fillna(default_time, inplace=True)
data.drop('Flight #',axis=1,inplace=True)
```
